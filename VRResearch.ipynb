```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned CSV
df = pd.read_csv('/content/IPQ_NASA_TLX_Cleaned.csv')

# Question titles from Google Form
nasa_titles = {
    1: 'How mentally demanding was the task?',
    2: 'How physically demanding was the task?',
    3: 'How hurried or rushed did you feel?',
    4: 'How successful do you feel you were at completing the task?',
    5: 'How hard did you have to work to accomplish the task?',
    6: 'How irritated, stressed, or annoyed were you while completing the task?'
}

# Loop through NASA questions
for qnum in range(1, 7):
    hand_col = f'NASA_Q{qnum}_Hand'
    ray_col = f'NASA_Q{qnum}_Ray'

    # Calculate stats
    mean_hand = df[hand_col].mean()
    median_hand = df[hand_col].median()
    mode_hand = df[hand_col].mode().values[0] if not df[hand_col].mode().empty else None
    std_hand = df[hand_col].std()

    mean_ray = df[ray_col].mean()
    median_ray = df[ray_col].median()
    mode_ray = df[ray_col].mode().values[0] if not df[ray_col].mode().empty else None
    std_ray = df[ray_col].std()

    # Print stats
    print(f"{hand_col} Statistics:")
    print(f"Mean: {mean_hand:.2f}")
    print(f"Median: {median_hand}")
    print(f"Mode: {mode_hand}")
    print(f"Standard Deviation: {std_hand:.2f}\n")

    print(f"{ray_col} Statistics:")
    print(f"Mean: {mean_ray:.2f}")
    print(f"Median: {median_ray}")
    print(f"Mode: {mode_ray}")
    print(f"Standard Deviation: {std_ray:.2f}\n")

    # --- Bar plot ---
    plt.figure(figsize=(8,6))
    bars = plt.bar([hand_col, ray_col], [mean_hand, mean_ray],
                   color=['skyblue', 'lightgreen'], edgecolor='black')
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.1, f'{height:.2f}', ha='center', fontsize=12)
    plt.title(f"Mean Scores: {nasa_titles[qnum]}", fontsize=16)
    plt.ylabel('Mean Score', fontsize=14)
    plt.ylim(0, 10)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

    # --- Boxplot + stripplot ---
    df_long = df[[hand_col, ray_col]].melt(var_name='Interaction', value_name='Score')
    sns.set(style='whitegrid', font_scale=1.2)
    plt.figure(figsize=(10,7))
    sns.boxplot(x='Interaction', y='Score', data=df_long, palette='pastel')
    sns.stripplot(x='Interaction', y='Score', data=df_long, color='black', size=8, jitter=True, alpha=0.7)
    plt.title(nasa_titles[qnum], fontsize=16)
    plt.xlabel('Interaction Type', fontsize=14)
    plt.ylabel('Score', fontsize=14)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned CSVs
df_ray = pd.read_csv('/content/Ray_Interactor_Cleaned.csv')
df_hand = pd.read_csv('/content/Hand_Interactor_Cleaned.csv')

# Get numeric scores
ray_scores = df_ray['Numeric_Score']
hand_scores = df_hand['Numeric_Score']

# Compute stats
mean_ray = ray_scores.mean()
median_ray = ray_scores.median()
mode_ray = ray_scores.mode().values[0] if not ray_scores.mode().empty else None
std_ray = ray_scores.std()

mean_hand = hand_scores.mean()
median_hand = hand_scores.median()
mode_hand = hand_scores.mode().values[0] if not hand_scores.mode().empty else None
std_hand = hand_scores.std()

# Print stats
print("Ray Interactor Stats:")
print(f"Mean: {mean_ray:.2f}")
print(f"Median: {median_ray}")
print(f"Mode: {mode_ray}")
print(f"Standard Deviation: {std_ray:.2f}\n")

print("Hand Interactor Stats:")
print(f"Mean: {mean_hand:.2f}")
print(f"Median: {median_hand}")
print(f"Mode: {mode_hand}")
print(f"Standard Deviation: {std_hand:.2f}")

# --- Bar plot of means ---
plt.figure(figsize=(8,6))
bars = plt.bar(['Ray Interactor', 'Hand Interactor'], [mean_ray, mean_hand],
               color=['skyblue', 'lightgreen'], edgecolor='black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.05, f'{height:.2f}', ha='center', fontsize=12)

plt.title('Average Correct Answers', fontsize=16)
plt.ylabel('Average Score (out of 6)', fontsize=14)
plt.ylim(0, 6)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# --- Boxplot of scores ---
# Prepare long format for seaborn
df_long = pd.DataFrame({
    'Interactor': ['Ray'] * len(ray_scores) + ['Hand'] * len(hand_scores),
    'Score': pd.concat([ray_scores, hand_scores], ignore_index=True)
})

plt.figure(figsize=(10,7))
sns.boxplot(x='Interactor', y='Score', data=df_long, palette='pastel')
sns.stripplot(x='Interactor', y='Score', data=df_long, color='black', size=8, jitter=True, alpha=0.7)

plt.title('Score Distributions by Interactor', fontsize=16)
plt.xlabel('Interaction Type', fontsize=14)
plt.ylabel('Correct Answers (out of 6)', fontsize=14)
plt.show()
```
